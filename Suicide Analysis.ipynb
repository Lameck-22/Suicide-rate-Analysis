## Yearly Sucide Rate

```python
suicide_df3= suicide_df2.groupby('year')['suicides_no'].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(suicide_df3['year'], suicide_df3['suicides_no'], marker='o', linestyle='-', color='b')
plt.title('Number of Suicides per Year')
plt.xlabel('Year')
plt.ylabel('Number of Suicides')
plt.grid(True)
plt.show()
```

```python
suicide_df4= suicide_df2.groupby('year')['suicides/100k pop'].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(suicide_df4['year'], suicide_df4['suicides/100k pop'], marker='o', linestyle='-', color='b')
plt.title('suicides/100k pop')
plt.xlabel('Year')
plt.ylabel('suicides/100k pop')
plt.grid(True)
plt.show()
```

```python
# Group by country and calculate the mean suicide rate per 100k population
country_suicide_rate = suicide_df2.groupby('country')['suicides/100k pop'].mean().reset_index()
# Sort values for better visualization and get top 40 countries
country_suicide_rate = country_suicide_rate.sort_values(by='suicides/100k pop', ascending=True).head(40)

# Number of countries
num_countries = len(country_suicide_rate)
# Create a color map
colors = plt.cm.viridis(np.linspace(0, 1, num_countries))
# Sort values for better visualization
country_suicide_rate = country_suicide_rate.sort_values(by='suicides/100k pop', ascending=True)

plt.figure(figsize=(20, 25))
plt.barh(suicide_df2['country'], suicide_df2['suicides/100k pop'], color=colors)
plt.title('Average Suicide Rate per 100k Population by Country')
plt.xlabel('Suicide Rate per 100k Population')
plt.ylabel('Country')
plt.grid(True, axis='x', linestyle='--', alpha=0.7)
plt.show()

```

```python
# Remove any commas and convert GDP to numeric
suicide_df2['gdp_for_year ($)2'] = suicide_df2['gdp_per_capita ($)'].replace({',': ''}, regex=True).astype(float)
```

```python
# Group by sex and calculate the average suicide rate
gender_suicide_rate = suicide_df2.groupby('sex')['suicides/100k pop'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=gender_suicide_rate, x='sex', y='suicides/100k pop', palette='Set2')
plt.title('Average Suicide Rate per 100k Population by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Suicide Rate per 100k Population')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group by sex and calculate the average suicide rate
generation_suicide_rate = suicide_df2.groupby('generation')['suicides/100k pop'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=generation_suicide_rate, x='generation', y='suicides/100k pop', palette='Set2')
plt.title('Average Suicide Rate per 100k Population by Generation')
plt.xlabel('Generation')
plt.ylabel('Average Suicide Rate per 100k Population')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=suicide_df2, x='gdp_per_capita ($)', y='suicides/100k pop')
#sns.regplot(x='gdp_per_capita ($)', y='suicides/100k pop', data=suicide_df2, scatter_kws={'s':100}, line_kws={'color':'red'}, ci=None)
#sns.scatterplot(suicide_df2['gdp_per_capita ($)'], suicide_df2['suicides/100k pop'], marker='o', linestyle='-', color='b')
plt.title('Number of Suicides per gdp/capita')
plt.xlabel('gdp_per_capita ($)')
plt.ylabel('Number of Suicides')
#plt.grid(True)
plt.show()
```

```python
# Group by age and calculate the average suicide rate
age_suicide_rate = suicide_df2.groupby('age')['suicides/100k pop'].mean().reset_index()
plt.figure(figsize=(8, 8))

# Plot pie chart
plt.pie(
    age_suicide_rate['suicides/100k pop'], 
    labels=age_suicide_rate['age'], 
    autopct='%1.1f%%',  # Show percentages
    startangle=140,  # Start angle for better visual effect
    colors=plt.cm.viridis(np.linspace(0, 1, len(age_suicide_rate)))
)

plt.title('suicides/100k vs age')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```
